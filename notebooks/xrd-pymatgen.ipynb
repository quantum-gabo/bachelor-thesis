# XRD pattern for CSH using pymatgen

```python
from pymatgen.core import Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
import scienceplots
# Set the style for the plots
plt.style.use(['science', 'notebook', 'no-latex'])
```

The `POSCAR` file I loaded here is the last one we relaxed using PBESol+IVWD functional. 

```python
# Loading my structure
csh_structure = Structure.from_file("../cluster-files/XRD/POSCAR")
```

```python
# Initialize XRD simulator (Cu KÎ±)
xrd = XRDCalculator(wavelength = "CuKa", symprec=0.5)
```

```python
# Get pattern data
pattern = xrd.get_pattern(csh_structure, two_theta_range=(5, 50), scaled = True)
```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(pattern.x, pattern.y, label="Simulated XRD")
plt.xlabel("2Î¸ (Â°)")
plt.ylabel("Intensity (a.u.)")
plt.title("XRD Pattern of C-S-H")
plt.ylim(0, 50)
plt.legend()
plt.grid(True)
plt.show()
```
